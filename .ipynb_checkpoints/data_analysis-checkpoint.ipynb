## Déplacement image

```python
def deplace (img , x_up = 0 , x_down = 0 , x_left = 0, x_right = 0):
    """Renvoi l'image déplacée suivant les 4 directions """
    
    px , py = np.shape(img)
    h = int (abs ( x_up - x_down ))
    v = int (abs ( x_right - x_left ))
    
    #Sous-fonctions de déplacement
    def up(img , x_up : int):
        deplaced_img = np.zeros ( (px , py) , dtype = float)
        for i in range(px - x_up):
            deplaced_img[i] = img[i + x_up]
        return deplaced_img
    
    def down(img , x_down: int):
        deplaced_img = np.zeros ( (px , py) , dtype = float)
        for i in range(x_down, px):
            deplaced_img[i] = img[i - x_down]
        return deplaced_img
    
    def left(img , x_left : int):
        deplaced_img = np.zeros ( (px , py) , dtype = float)
        for i in range(px):
            for j in range (py - x_left):
                deplaced_img[i][j] = img[i][j + x_left]
        return deplaced_img
    
    def right(img , x_right : int):
        deplaced_img = np.zeros ( (px , py) , dtype = float)
        for i in range(px):
            for j in range (x_right, py):
                deplaced_img[i][j] = img[i][j - x_right]
        return deplaced_img
    
    #Choix du cadran
    if x_up >= x_down and x_right >= x_left :
        return up  (right(img,v),h)
    
    if x_up <= x_down and x_right >= x_left :
        return down(right(img,v),h)
    
    if x_up >= x_down and x_right <= x_left :
        return up  (left (img,v),h)
    
    if x_up <= x_down and x_right <= x_left :
        return down(left (img,v),h)
    
    
```

## Contraste 

```python
def change_contrast (img : np.ndarray):
    
    px , py = np.shape(img)
    c_img = np.copy(img)
    for i in range(px):
        for j in range (py):
            e = img[i][j]
            if e < 0.8:
                c_img[i][j] = e - 0.3
    return c_img
```

## Data augmentation

```python
def dust (img):
    """Renvoie la "matrice bruitée" d'une matrice image d'entrée"""
    
    def gen_dust(size): 
        d = np.zeros((size,size))
        for i in range(size):
            for j in range(size):
                if r.random()>0.5:
                    d[i][j] = 0.5
        return d
    
    px , py = np.shape(img)
    dimg = np.copy(img)
    for i in range (px):
        for j in range(py):
            if r.random() < 0.05 :
                size = r.randrange(5,8)
                d = gen_dust(size)
                for m in range(size):
                    for n in range(size):
                        if dimg[(i+m)%px][(j+n)%py] < 0.6:
                            dimg[(i+m)%px][(j+n)%py] = d[m][n]
    return dimg
```

```python
def noise (M):
    px , py = np.shape(M)
    Mnoise = np.copy(M)
    for i in range (px):
        for j in range (py):
            if M[i][j] < 0.9:
                Mnoise[i][j] = r.random()
    return  Mnoise 

def modif_img (img):
    """Renvoie la matrice modifiée à partir d'une matrice de sortie"""
    
    #récupération des caractéristiques de l'image à modifier 
    px , py = np.shape(img)
    l = min(px , py)
    
    #Paramètres randomisés pour la modification 
    crop_frame = r.randint ( l//2 ,(l//4)*3 )
    i_start = r.randint(0 , px - crop_frame)
    j_start = r.randint(0 , py - crop_frame)
    gamma   = (r.random() * 3) + 2
    sigma   = (r.random() * 3) + 1
    theta   = r.randint(0,360)
    
    #Traitement image
    m_img = t.rotate  (img , theta )
    m_img = noise(m_img)
    m_img = util.crop ( m_img , ( (i_start ,px - i_start - crop_frame),(j_start ,py - j_start - crop_frame)) , copy=True)
    m_img = t.resize  ( m_img , (px,py))
    m_img = gaussian  ( m_img , sigma)
    #m_img = exp.adjust_gamma ( m_img , gamma = gamma , gain = 1 )
    
    return m_img


def create_mix_BDD (dir_name : str):
    
    output_dir_name = dir_name + '_mix' 
    
    if not os.path.exists(f'{workpath}\{output_dir_name}'):
        os.makedirs(f'{workpath}\{output_dir_name}')
        
    if len(os.listdir( f'{workpath}\{output_dir_name}')) ==0 :
    
        L = os.listdir(f'{workpath}\{dir_name}')
        i = 0
        for name in L:
            img = io.imread(f'{workpath}\{dir_name}\{name}', as_gray=True)
                            
            if i%2 == 0 :
                plt.imsave(f'{workpath}\{output_dir_name}\{name}', img , cmap = 'gray' , format = 'jpg')
            else:
                m_img = modif_img(img)
                plt.imsave(f'{workpath}\{output_dir_name}\{name}', m_img , cmap = 'gray' , format = 'jpg')
            i += 1
            
def create_fullmodif_BDD (dir_name : str):
     
    output_dir_name = dir_name + '_fullmodif' 
    
    if not os.path.exists(f'{workpath}\{output_dir_name}'):
        os.makedirs(f'{workpath}\{output_dir_name}')
        
    if len(os.listdir( f'{workpath}\{output_dir_name}')) ==0 :
    
        L = os.listdir(f'{workpath}\{dir_name}')
        for name in L:
            img = io.imread(f'{workpath}\{dir_name}\{name}', as_gray=True)
            m_img = modif_img(img)                
            plt.imsave(f'{workpath}\{output_dir_name}\{name}', m_img , cmap = 'gray' , format = 'jpg')
```

## Fonctions modif alternatives

```python
def data_augmentation (input_file_path : str , NB_img2take : int):
    """Augmente le dataset renseigné par le chemin, d'un nombre = à NB_img2take"""
    
    #Récupération des données des images sources
    L = os.listdir(input_file_path)
    px , py = np.shape(io.imread(f'{input_file_path}\{L[0]}' , as_gray = True))
    l = min(px , py)

    #if not os.path.exists(f'{workpath}\{output_file}'):
    #    os.makedirs(f'{workpath}\{output_file}')
    
    for c in range(NB_img2take):
        
        random_index = r.randint(0 , len(L) - 1)
        file_name = f'{L[random_index][:-4]}-MODIF{c}.jpg'
        random_img = io.imread (f'{input_file_path}\{L[random_index]}' , as_gray = True)
        
        #Paramètres randomisés pour la modification
        crop_frame = r.randint ( l//2 ,(l//4)*3 )
        i_start = r.randint(0 , px - crop_frame)
        j_start = r.randint(0 , py - crop_frame)
        gamma   = (r.random() * 3) + 2
        sigma   = (r.random() * 3) + 1
        
        #Traitement image
        m_img  = util.crop ( random_img , ( (i_start ,px - i_start - crop_frame),(j_start ,py - j_start - crop_frame)) , copy=True)
        m_img = t.resize ( m_img , (px,py))
        m_img = exp.adjust_gamma ( m_img , gamma , gain = 1)
        m_img = gaussian ( m_img , sigma) 
        
        #Sauvegarde dans le fichier source
        plt.imsave(f'{input_file_path}\{file_name}', m_img , cmap = 'gray' , format = 'jpg')
```

```python
def fill_file (input_file_path : str , NB_img2take : int , output_file_path : str ):
    
    #Récupération des données des images sources
    L = os.listdir(input_file_path)
    px , py = np.shape(io.imread(f'{input_file_path}\{L[0]}' , as_gray = True))
    l = min(px , py)

    if not os.path.exists(f'{output_file_path}'):
        os.makedirs(f'{output_file_path}')
    
    for c in range(NB_img2take):
        
        random_index = r.randint(0 , len(L) - 1)
        file_name = f'{L[random_index][:-4]}-MODIF ({c}).jpg'
        random_img = io.imread (f'{input_file_path}\{L[random_index]}' , as_gray = True)
        
        #Paramètres randomisés pour la modification
        crop_frame = r.randint ( l//2 ,(l//4)*3)
        i_start = r.randint(0 , px - crop_frame)
        j_start = r.randint(0 , py - crop_frame)
        gamma   = (r.random() * 3) + 2 
        sigma   = (r.random() * 3) + 1
        
        #Traitement image
        m_img  = util.crop ( random_img , ( (i_start ,px - i_start - crop_frame),(j_start ,py - j_start - crop_frame)) , copy=True)
        m_img = t.resize ( m_img , (px,py))
        m_img = exp.adjust_gamma ( m_img , gamma , gain = 1)
        m_img = dust(m_img)
        m_img = gaussian ( m_img , sigma) 
        
        #Sauvegarde dans le fichier source
        plt.imsave(f'{output_file_path}\{file_name}', m_img , cmap = 'gray' , format = 'jpg')
```
