```python
def create_dust (N : int , size : int):
    # Cration dossier contenant les 'dust'
    if not os.path.exists(workpath+'\dust'):  
        os.makedirs(workpath+'\dust')
    if not os.path.exists(workpath+'\dust'+ f'\{size}'):  
        os.makedirs(workpath+'\dust'+ f'\{size}')
    
    #Cration lignes par lignes des 'dust' , + nom de sauvegarde
    n = 0 
    while n < N:
        d_name = ''
        d = np.zeros((size,size) , dtype = float)
        for i in range(size):
            d_name = d_name + str(i+1)
            for j in range(size):
                if r.random() > 0.5 :
                    d[i][j] = 0.5
                    d_name = d_name + 'b'
                else:
                    d_name = d_name + 'd'
        d_name = d_name + '.jpg'
        n+=1
        
        # on ne sauvegarde que les 'dust' nouvelles
        if d_name not in os.listdir(workpath+'\dust'+ f'\{size}'):
            io.imsave(workpath+'\dust'+ f'\{size}\{d_name}' , d, as_gray = True)
    
```

```python
def insert_dust (img):
    
    def gen_dust(size):
        d = np.zeros((size,size))
        for i in range(size):
            for j in range(size):
                if r.random()>0.5:
                    d[i][j] = 0.5
        return d
    
    px , py = np.shape(img)
    dimg = np.copy(img)
    for i in range (px):
        for j in range(py):
            if r.random() < 0.05 :
                size = r.randrange(5,8)
                d = gen_dust(size)
                for m in range(size):
                    for n in range(size):
                        if dimg[(i+m)%px][(j+n)%py] < 0.6:
                            dimg[(i+m)%px][(j+n)%py] = d[m][n]
    return dimg

            
            
```

```python
def noise (M):
    px , py = np.shape(M)
    Mnoise = np.copy(M)
    for i in range (px):
        for j in range (py):
            if M[i][j] < 0.9:
                Mnoise[i][j] = r.random()
    return  Mnoise 
```

```python
def bb (M):
    min_value = min (M.ravel())
    px , py = np.shape(M)
    Mz = np.zeros ((px,py) , dtype = np.float16)
    for i in range(px):
        for j in range (py):
            if M[i][j] > 2*min_value:
                Mz[i][j] = 1
    return (Mz)
```
